# Ungraded Lab: Build a Multi-output Model

In this lab, we'll show how you can build models with more than one output. The dataset we will be working on is available from the [UCI Machine Learning Repository](https://archive.ics.uci.edu/ml/datasets/Energy+efficiency). It is an Energy Efficiency dataset which uses the bulding features (e.g. wall area, roof area) as inputs and has two outputs: Cooling Load and Heating Load. Let's see how we can build a model to train on this data.

## Imports

```python
try:
  # %tensorflow_version only exists in Colab.
  %tensorflow_version 2.x
except Exception:
  pass;

import tensorflow as tf;
import numpy as np;
import matplotlib.pyplot as plt;
import pandas as pd;
from tensorflow.keras.models import Model;
from tensorflow.keras.layers import Dense, Input;
from sklearn.model_selection import train_test_split;
```

## Utilities

We define a few utilities for data conversion and visualization to make our code more neat.

```python
def format_output(data: pd.DataFrame) -> tuple:
    y1 = data.pop('Y1');
    y1 = np.array(y1);
    y2 = data.pop('Y2');
    y2 = np.array(y2);
    return y1, y2;


def norm(x: pd.DataFrame) -> pd.DataFrame:
    return (x - train_stats['mean']) / train_stats['std'];


def plot_diff(y_true: np.ndarray, y_pred: np.ndarray, title: str='') -> None:
    plt.scatter(y_true, y_pred);
    plt.title(title);
    plt.xlabel('True Values');
    plt.ylabel('Predictions');
    plt.axis('equal');
    plt.axis('square');
    plt.xlim(plt.xlim());
    plt.ylim(plt.ylim());
    plt.plot([-100, 100], [-100, 100]);
    plt.show();


def plot_metrics(metric_name: str, title: str, ylim: int=5) -> None:
    plt.title(title);
    plt.ylim(0, ylim);
    plt.plot(history.history[metric_name], color='blue', label=metric_name);
    plt.plot(
        history.history[f'val_{metric_name}'],
        color='green',
        label=f'val_{metric_name}'
    );
    plt.show();
```

## Prepare the Data

We download the dataset and format it for training.

```python
# Specify data URI
URI: str = './data/ENB2012_data.xlsx';

# Use pandas excel reader
df: pd.DataFrame = pd.read_excel(URI);
df = df.sample(frac=1).reset_index(drop=True);

# Split the data into train and test with 80 train / 20 test
train: pd.DataFrame;
test: pd.DataFrame;
train, test = train_test_split(df, test_size=0.2);
train_stats: pd.DataFrame = train.describe();

# Get Y1 and Y2 as the 2 outputs and format them as np arrays
train_stats.pop('Y1');
train_stats.pop('Y2');
train_stats = train_stats.transpose();
train_Y: tuple = format_output(train);
test_Y: tuple = format_output(test);

# Normalize the training and test data
norm_train_X: pd.DataFrame = norm(train);
norm_test_X: pd.DataFrame = norm(test);
```

## Build the Model

Here is how we'll build the model using the functional syntax. Notice that we can specify a list of outputs (i.e. `[y1_output, y2_output]`) when we instantiate the `Model()` class.

```python
# Define model layers.
input_layer: tf.keras.layers.Layer = Input(shape=(len(train.columns),));
first_dense: tf.keras.layers.Layer = Dense(
    units='128', activation='relu'
)(input_layer);
second_dense: tf.keras.layers.Layer = Dense(
    units='128', activation='relu'
)(first_dense);

# Y1 output will be fed directly from the second dense
y1_output: tf.keras.layers.Layer = Dense(
    units='1', name='y1_output'
)(second_dense)
third_dense: tf.keras.layers.Layer = Dense(
    units='64', activation='relu'
)(second_dense);

# Y2 output will come via the third dense
y2_output: tf.keras.layers.Layer = Dense(
    units='1', name='y2_output'
)(third_dense);

# Define the model with the input layer and a list of output layers
model: tf.keras.Model = Model(
    inputs=input_layer, outputs=[y1_output, y2_output]
);

print(model.summary());
```

## Configure parameters

We specify the optimizer as well as the loss and metrics for each output.

```python
# Specify the optimizer, and compile the model with loss functions for both outputs
optimizer: tf.keras.optimizers.Optimizer = tf.keras.optimizers.SGD(lr=0.001)
model.compile(
    optimizer=optimizer,
    loss={'y1_output': 'mse', 'y2_output': 'mse'},
    metrics={
        'y1_output': tf.keras.metrics.RootMeanSquaredError(),
        'y2_output': tf.keras.metrics.RootMeanSquaredError()
    }
);
```

## Train the Model

```python
# Train the model for 500 epochs
history: tf.keras.callbacks.History = model.fit(
    norm_train_X,
    train_Y,
    epochs=500,
    batch_size=10,
    validation_data=(norm_test_X, test_Y)
);
```

## Evaluate the Model and Plot Metrics

```python
# Test the model and print loss and mse for both outputs
loss: np.ndarray;
Y1_loss: np.ndarray;
Y2_loss: np.ndarray;
Y1_rmse: np.ndarray;
Y2_rmse: np.ndarray;
loss, Y1_loss, Y2_loss, Y1_rmse, Y2_rmse = model.evaluate(
    x=norm_test_X, y=test_Y
);
print(f"Loss = {loss}, Y1_loss = {Y1_loss}, Y1_mse = {Y1_rmse}, Y2_loss = {Y2_loss}, Y2_mse = {Y2_rmse}");
```

```python
# Plot the loss and mse
Y_pred: np.ndarray = model.predict(norm_test_X);
plot_diff(test_Y[0], Y_pred[0], title='Y1');
plot_diff(test_Y[1], Y_pred[1], title='Y2');
plot_metrics(
    metric_name='y1_output_root_mean_squared_error', title='Y1 RMSE', ylim=6
);
plot_metrics(
    metric_name='y2_output_root_mean_squared_error', title='Y2 RMSE', ylim=7
);
```
